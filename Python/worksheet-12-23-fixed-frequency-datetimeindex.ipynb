### **pandas.date_range**
Return a fixed frequency Datetimeindex

```python
y = pd.date_range(start = "2021-01-01", end = "2021-01-31")
y
```

```python
# Specifying the number of periods
y = pd.date_range(start="2021-01-01", end = "2021-01-31", periods=50)
y
```

```python
# Specifying only start date along with no. of periods
y = pd.date_range(start="2021-01-01", periods=100)
y
```

```python
# Specifying only end date along with no. of periods
y = pd.date_range(end="2021-01-31", periods=50)
y
```
