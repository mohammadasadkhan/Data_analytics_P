# Customising the stock data using Time period
We have two ways to fetch the data using yf.download() function
1. Either Use period parameter or use start and end 
2. Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max

We are going to explore both the arguments here

```python
#intsalling the library
!pip install yfinance
```

```python
#importing the library
import yfinance as yf
```

```python
#fixing the ticker
ticker="tsla"
```

# Fetching the data using Start and End Date 
We can use the dates by defining in string format, using the sequence #YYYY-MM-DD

```python
#setting the start date
start="2021-04-07"
```

```python
#setting the end date
end="2022-04-07"
```

```python
df=yf.download(ticker,start=start,end=end)
```

```python
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(18, 6), dpi=80)
plt.plot(df.Close)
plt.show()
```
