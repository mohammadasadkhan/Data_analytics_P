# **Scatterplot**

A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. If the points are coded, one additional variable can be displayed.

```python
d = [[14.2,215],
[16.4,325],
[11.9,185],
[15.2,332],
[18.5,406],
[22.1,522],
[19.4,412],
[25.1,614],
[23.4,544],
[18.1,421],
[22.6,445],
[17.2,408]]
```

```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data
df = pd.DataFrame(data=d, columns = ["Temp","Sales"])
```

```python
# Get the head()
df.head()
```

```python
# Simple plot
df.plot()
```

```python
# Plot scatter using kind="scatter" and defining x and y columns
df.plot(kind = "scatter", x = "Temp",y = "Sales")
```

```python
# Change the color
df.plot(kind = "scatter", x = "Temp",y = "Sales", c="r")
```

```python
# Change the marker
df.plot(kind = "scatter", x = "Temp",y = "Sales", c="r", marker="x")
```

```python
# Change the size od scatter using a parameter
df.plot(kind = "scatter", x = "Temp",y = "Sales",c="r",marker="x", s=50, figsize=(12,5))
```
