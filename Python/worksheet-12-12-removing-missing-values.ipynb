# Importing the IPL 2022 Auction Dataset

```python
df=pd.read_csv("../input/2022-ipl-auction-dataset/ipl_2022_dataset.csv",index_col= "Unnamed: 0")
```

# Checking the info() of the dataframe

```python
#this will help us identify the null values
df.info()
```

# Producing a dataframe checking isna()

```python
# using isna() this will return us boolean dataframe, true at null value and false at some values
df.isna()
```

# Summing up all the null values()

```python
# using sum() along with isna()
# this will return the total number of null values in specific columns
df.isna().sum()
```

# Producing a dataframe of null values present in a specific column

```python
# dataframe of the values where df["COST IN ₹ (CR.)"] is a null values 
df[df["COST IN ₹ (CR.)"].isna()]
```

# Checking the value_counts() along with null values

```python
# using value_counts() to know what are the diffrent values we have in the column and the count of occurrences 
df["COST IN ₹ (CR.)"].value_counts(dropna=False)
```

# Calculating the mean of a column by skipping the null values

```python
# calculating the mean by skipping the null values 
# if we will not skill the null values the mean will be corrupted 
df["COST IN ₹ (CR.)"].mean(skipna=True)
```

# Dropping null values using any & all arguments

```python
df.shape
```

```python
# this will drop the rows which contains even one null value
df.dropna(axis=1,how="any").shape
```

```python
# this will drop the rows which contains all the elements as null values
df.dropna(axis=0,how="any").shape
```

```python
# this will drop the columns as the axis is defined as 1
# if any element in the column is null values the column will be dropped
# we use 0 for rows and 1 for columns
df.dropna(axis=1,how="all").shape
```

```python
# the column will be dropped only if all the elements are null values
df.dropna(axis=0,how="all").shape
```

```python
df.shape
```

# Dropping null values with a thresh

```python
#thresh - thresh takes integer value which tells minimum amount of na values to drop.
df.dropna(axis=1,thresh=400).shape
```

```python
#thresh - thresh takes integer value which tells minimum amount of na values to drop.
df.dropna(axis=1,thresh=700).shape
```
