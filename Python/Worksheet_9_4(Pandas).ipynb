# **Index Operator iloc (Position based Indexing)**

```python
# Get olympic data from kaggle
# Read the data
df1 = pd.read_excel("/content/drive/MyDrive/Project/Practice set/Medals.xlsx")
```

```python
df1
```

```python
# Show the head
df1.head()
```

```python
# Show the tail
df1.tail()
```

```python
# Show info of the dataframe
df1.info()
```

```python
# Fetching the first row using iloc
df1.iloc[0]
```

```python
# Checking the type of the output
type(df1.iloc[0])
```

```python
# Fetching the last row using negative indexing
df1.iloc[-1]
```

```python
# Fetching first 5 rows using iloc inplace of head
df1.iloc[:5]
```

```python
# Fetching last 5 rows using iloc inplace of tail
df1.iloc[-5:]
```

```python
# Fetching 5 consecutive rows using iloc
df1.iloc[5:9]
```

```python
# Fetching 3 different rows from differect locations using iloc
df1.iloc[[25,46,57],]
```

```python
# Fetching only the first row and first column using iloc
df1.iloc[0,0]
```

```python
# Fetching the first row and two non-consecutive columns
df1.iloc[0,[3,4]]
```

```python
# Fetching 3 different rows and 3 different columns
df1.iloc[[25,46,57],[0,2,4]]
```

```python
# Fetching 3 consecutive rows and 3 cosecutive columns
df1.iloc[:3,:3]
```

```python
# Compare equal of same column, fetching one by attribute and other one by iloc
df1.Gold
```

```python
df1.Gold.equals(df1.iloc[:,2])
```
