# **Visualization with Matplotlib**

```python
# Import pandas library
import pandas as pd
```

```python
# Read the data
df = pd.read_csv("/content/drive/MyDrive/Project/Practice set/RELIANCE.NS.csv", index_col="Date")
```

```python
# Show the data head
df.head()
```

```python
# Get the info of dataframe
df.info()
```

```python
# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Simply plotting the whole dataframe by using .plot method
df.plot()
plt.show()
```

# **Note**

plt.show() is used when python is being written over a terminal. It is not required while writing the code over a notebook

```python
# Increase the size of the plot using figsize
# The figsize argument will take a tuple as input
df.plot(figsize=(15,5))
```

```python
df.plot(subplots = True, figsize=(15,10))
```

```python
# Here all the plots are sharing same axes which can be changed
# Set sharex-False which is True by default
# Playaround sharey too
df.plot(subplots = True, figsize=(15,12), sharex = False)
```

```python
# Plot for only one column
df.Close.plot(figsize=(15,5))
```
