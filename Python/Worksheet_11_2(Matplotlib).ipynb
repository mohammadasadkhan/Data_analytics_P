# **Customisation of Plots**

```python
# Import libraries pandas and matplotlib
import pandas as pd
import matplotlib as plt
```

```python
# Read the data using pandas
df = pd.read_csv("/content/drive/MyDrive/Project/Practice set/RELIANCE.NS.csv", index_col="Date")
```

```python
df.head()
```

```python
# Plot a lineplot using matplotlib
df.plot(subplots=True, figsize=(12,10))
```

```python
df.info()
```

```python
# Plot fro a single column
df.Low.plot(figsize=(12,5))
```

```python
# Change font size by fontsize argument in .plot() method
df.Low.plot(figsize=(12,5), fontsize=12)
```

# [**matplotlib.pyplot.plot**](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
# Change color by c argumnet in .plot() method
df.Low.plot(figsize=(12,5),c="r")
```

```python
# Change linestyle by linestyle argumnet in .plot() method
df.Low.plot(figsize=(12,5),c="r", linestyle=":")
```

```python
# Change the title of the plot using .title() method
df.Low.plot(figsize=(12,5),c="r", linestyle=":", title="Reliance Stock Data")
```

```python
# Insert legend using .legend() method, change location and fontsize
df.Low.plot(figsize=(12,5),c="r", linestyle=":", title="Reliance Stock Data", legend = True)
```

```python
# Put some lebels on thh edges
# Us .xlabel() and .ylabel() methods, customize the fontsize too
df.Low.plot(figsize=(12,5),c="r", linestyle=":", ylabel = "Price")
```

```python
# Show the gride using .grid() method
df.Low.plot(figsize=(12,5),c="r", linestyle=":", grid=True)
```

```python
# Set x axes and y axes limits using xlim and ylim parameters of .plot() method
df.Low.plot(figsize=(12,5),c="r", linestyle=":", grid=True, ylim=[2200,2700])
```
