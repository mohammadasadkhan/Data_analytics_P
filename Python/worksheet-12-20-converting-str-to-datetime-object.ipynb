## **Different Example of Converting String to Datetime Object**

```python
pd.to_datetime("2022-04-27")
```

```python
pd.to_datetime("2022-04-27 10")
```

```python
pd.to_datetime("20220427 10:29:20")
```

```python
pd.to_datetime("2022 04 27 10:29:20")
```

```python
pd.to_datetime("2022 April 27 10:29:20")
```

```python
pd.to_datetime("April 2022 27 10:29:20")
```

errors : {'ignore','raise','coerce'}, default 'raise'
* If 'raise', then invalid parsing will raise an exception.
* If 'coerce', then invalid parsing will be set as Nat.
* If 'igonre', then invalid parsing will return the input.

```python
pd.to_datetime(["2022 27 April 10:29:20","testing"],errors="coerce")
```
