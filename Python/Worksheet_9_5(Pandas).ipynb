# **Index Operator loc (label based Indexing)**

```python
# Import the library
import pandas as pd
```

```python
# Read the csv Data
df = pd.read_csv("/content/drive/MyDrive/Project/Practice set/RELIANCE.NS.csv",index_col="Date")
```

```python
# Show the head()
df.head()
```

```python
df.info()
```

Note: To fetch via label based indexing it is adviced that the index of the the data that you are using must have labels, In case it dosen't please set another column an index while reading the data from pandas

```python
# Using loc fetch the data of one particular date
df.loc["2022-09-05"]
```

```python
# Using loc fetch the data of one particular date and one column name
df.loc["2022-09-05","Volume"]
```

```python
df.loc["2022-09-05","High"]
```

```python
# Using loc fetch the data of two particular date and two column name
df.loc[["2022-09-05","2022-09-07","2022-09-08"],["High","Volume"]]
```

```python
# Slicing dates data using loc
df.loc[:"2022-09-07"]
```

```python
df.tail()
```

```python
# Slicing dates data using loc definig the starting poin
df.loc["2023-08-04":]
```

```python
# Slicing dates data using loc definig the ending point.
df.loc[:"2022-10-04"]
```

```python
# Slicing dates data using loc definig both the starting and ending point
df.loc["2020-09-23":"2022-10-23"]
```

```python
# Slicing both in columns and rows using loc
df.loc[["2020-09-23","2020-10-23"],["Open","High"]]
```
