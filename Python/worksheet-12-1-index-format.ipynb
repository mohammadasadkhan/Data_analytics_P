## **Changing the header**

As the default behavior of the parameter header, the first row is selected as header (header=0) We can change the header by passing the index value or we can set it as None

```python
# Changing the heder to index 1
df2=pd.read_csv("/kaggle/input/covid19-in-india/covid_19_india.csv",header=1)
df2
```

```python
# Changing the heder to index None
df3=pd.read_csv("/kaggle/input/covid19-in-india/covid_19_india.csv",header=None)
df3
```
